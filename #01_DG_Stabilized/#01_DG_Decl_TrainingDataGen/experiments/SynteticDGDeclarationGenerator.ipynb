```python
def get_synonyms(term, synonyms_csv):
    """
    Liefert eine Liste von Synonyme für einen Begriff.

    Args:
        term: Der Begriff, für den Synonyme gesucht werden sollen.
        synonyms_csv: Der Dateipfad zur CSV-Datei mit den Synonyme.

    Returns:
        Eine Liste von Synonyme für den Begriff.
    """

    synonyms = []
    with open(synonyms_csv, "r") as f:
        reader = pd.read_csv(f,delimiter='|')
        for row in reader.itertuples():
            if row[1] == term:
                synonyms.append(row[1])
    return synonyms

def lemmatize(text):
    """
    Lemmatisiert einen Text.

    Args:
        text: Der Text, der lemmatisiert werden soll.

    Returns:
        Der lemmatisierte Text.
    """

    lemmatizer = WordNetLemmatizer()
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

def paraphrase(text, synonyms):
    """
    Paraphrasiert einen Text durch Ersetzen von Wörtern durch Synonyme.

    Args:
        text: Der Text, der paraphrasiert werden soll.
        synonyms: Die Liste von Synonyme, die verwendet werden sollen.

    Returns:
        Der paraphrasierte Text.
    """

    words = text.split()
    new_words = []
    for word in words:
        if word in synonyms:
            new_words.append(synonyms[word])
        else:
            new_words.append(word)
    return " ".join(new_words)

def bleu_score(original, paraphrase):
    """
    Berechnet die BLEU-Score für einen Originaltext und eine Paraphrase.

    Args:
        original: Der Originaltext.
        paraphrase: Die Paraphrase.

    Returns:
        Die BLEU-Score.
    """

    return sentence_bleu([original], paraphrase)
```

```python
# Beispielbegriff und Synonyme
term = "Hydrogen Peroxide"
synonyms_csv = "drive/MyDrive/Colab Notebooks/SynteticDGDeclarationGenerator/synonyms.txt"

# Synonyme abrufen
synonyms = get_synonyms(term, synonyms_csv)
print(synonyms)

# Beispielformulierungen
formulierungen = [
    "UN1202|Diesel fuel, Class 3, PG III, Conditioned for safe transport|STABILIZED",
    "UN1203|Gasoline, Class 3, PG II, Stabilized with proprietary additives|STABILIZED",
    "UN1263|Paint, flammable, Class 3, PG III, Formulated to minimize fire risk"]

# Paraphrasierte Formulierungen generieren
variants = []
for formulierung in formulierungen:
    sample = paraphrase(formulierung, synonyms)
    variants.append(sample)

# BLEU-Scores berechnen
bleu_scores = []
for original, variant in zip(formulierungen, variants):
    score = bleu_score(original, variants)
    bleu_scores.append(score)

# Ausgabe
print("Originale Formulierungen:")
for formulierung in formulierungen:
    print(formulierung)

print("Paraphrasierte Formulierungen:")
for variant in variants:
    print(variant)

print("BLEU-Scores:")
for bleu_score in bleu_scores:
    print(bleu_score)
```

```python
# Lade das LLM-Modell
model = transformers.AutoModelForCausalLM.from_pretrained("google/bigbird-roberta-base")
tokenizer = transformers.AutoTokenizer.from_pretrained("google/bigbird-roberta-base")

# Generiere paraphrasierte Varianten
def paraphrase(text, model, tokenizer, num_return_sequences=2):
    """
    Paraphrasiert einen Text unter Verwendung eines LLMs.

    Args:
        text: Der Text, der paraphrasiert werden soll.
        model: Das LLM-Modell.
        tokenizer: Der Tokenizer für das LLM-Modell.
        num_return_sequences: Die Anzahl der paraphrasierten Varianten, die generiert werden sollen.

    Returns:
        Eine Liste von paraphrasierten Varianten des Textes.
    """

    input_ids = tokenizer(text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids, max_length=len(input_ids) + 50, num_beams=5, num_return_sequences=num_return_sequences)
    vlst = []
    for output in outputs:
        var = tokenizer.decode(output, skip_special_tokens=True)
        vlst.append(var)
    return vlst

# Beispielbegriff
term = "Haus"

# Paraphrasierte Varianten generieren
variants = paraphrase(term, model, tokenizer)

def filter_variants(variants):
    """
    Filtert eine Liste von paraphrasierten Varianten.

    Args:
        variants: Eine Liste von paraphrasierten Varianten.

    Returns:
        Eine Liste von gefilterten paraphrasierten Varianten.
    """

    filtered_variants = []
    for variant in variants:
        if variant != term and len(variant.split()) >= 2:
            filtered_variants.append(variant)
    return filtered_variants

variants = filter_variants(variants)

# Ausgabe
print("Originaler Begriff:")
print(term)

print("Paraphrasierte Varianten:")
for variant in variants:
    print(variant)
```

```python
import pandas as pd
import ast
import os

def list_to_dataframe(input_text, un_code, label):
    try:
        input_list = ast.literal_eval(input_text)
    except:
        print(f"Failed to parse: {input_text}")
        return pd.DataFrame()

    columns = ["UN_CODE", "DG_DECLARATION", "LABEL"]
    df = pd.DataFrame(columns=columns)
    for dg_decl in input_list:
        new_row = pd.DataFrame([[un_code, dg_decl, label]], columns=columns)
        df = pd.concat([df, new_row], ignore_index=True)
    return df

def generate_all_variants(synonyms_csv, prompt_template, model, parameters, terms_to_generate=[]):
    csv_path = 'all_variants.csv'
    if os.path.exists(csv_path):
        master_df = pd.read_csv(csv_path)
    else:
        master_df = pd.DataFrame(columns=["UN_CODE", "DG_DECLARATION", "LABLE"])

    with open(synonyms_csv, "r") as f:
        df = pd.read_csv(f, delimiter='|')

    if not terms_to_generate:
        terms_to_generate = df['TERM'].tolist()

    for term in terms_to_generate:
        if 'UN_CODE' in master_df.columns and term in master_df['UN_CODE'].unique():
            print(f"Skipping {term}, already exists.")
            continue

        term_row = df[df['TERM'] == term].iloc[0]
        synonyms = term_row['SYNONYMES']
        un_code = term_row['UNCODE']

        prompt = prompt_template.replace("<Target>", term)
        prompt = prompt.replace("<Synonyms>", synonyms)

        try:
            res = generateSamples(prompt, model, parameters).text
            res = res.strip("`")
            res = res.strip("DG_declarations = ")
            df_samples = list_to_dataframe(res, un_code, 'TO BE STABILIZED')
            master_df = pd.concat([master_df, df_samples], ignore_index=True)
        except Exception as e:
            print(f"Failed for term {term}: {e}")

        master_df.to_csv(csv_path, index=False)

# Assuming prompt_template, model, and parameters are already defined
generate_all_variants("drive/MyDrive/Colab Notebooks/SynteticDGDeclarationGenerator/synonyms.txt", prompt_template, model, parameters)
```
